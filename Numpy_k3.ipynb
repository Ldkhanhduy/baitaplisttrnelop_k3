```python
import numpy as d
```

Cau 1

```python
n=abs(int(input('nhap so phan tu cho vector x:')))
x=d.random.randint(-10,10,n)
print('x=',x)
```

Cau 2

```python
x1=x.copy()
for i in range(len(x1)):
  if x1[i]<0:
    x1[i]*=(-1)
print('sin(x)=',d.sin(x1))
print('cos(x)=',d.cos(x1))
print('sqrt(x)=',d.sqrt(x1))
print('log(x)=',d.log(x1))
```

Cau 3

```python
print('sin(x)=',d.round_(d.sin(x1),1,None))
print('cos(x)=',d.round_(d.cos(x1),1,None))
print('sqrt(x)=',d.round_(d.sqrt(x1),1,None))
print('log(x)=',d.round_(d.log(x1),1,None))
```

Cau 4

```python
m=abs(int(input('nhap so hang cho ma tran:')))
n=abs(int(input('nhap so cot cho ma tran:')))
matran=d.random.randint(-10,10,m*n)
matran=d.reshape(matran,(m,n))
print(matran)
matranx=matran.copy()
print(matranx)
for i in range(len(matranx)):
  for j in range(len(matranx[i])):
    if matranx[i,j]<0:
      matranx[i,j]*=(-1)
print('sin(x)=',d.round_(d.sin(matranx),3,None))
print('cos(x)=',d.round_(d.cos(matranx),3,None))
print('sqrt(x)=',d.round_(d.sqrt(matranx),3,None))
print('log(x)=',d.round_(d.log(matranx),3,None))
```
